**Data limitation — updated finding:** Investigated WHY Tomato and Rice have partial month coverage by checking Year+Month combinations directly:
- Rice: ALL data is from 2025 only (April, May, June) — zero records from 2023 or 2024
- Tomato: ALL data is from 2023 only (June through November) — zero records from 2024 or 2025
- Onion, Potato: confirmed full 12-month coverage (needs verification across all 3 years, not just month count)
- Wheat: 9-month coverage (needs similar year-by-year verification)

Conclusion: this is a dataset construction/collection artifact — different crops were evidently scraped/tracked during different, non-overlapping time windows — NOT genuine crop seasonality. Rice and Tomato's monthly price figures reflect a single year's snapshot each, not a multi-year seasonal average. This means: (1) true year-over-year seasonality analysis is only valid for crops with multi-year coverage, (2) Rice/Tomato price prediction models (Day 8+) cannot be validated across multiple years and should be scoped/documented accordingly, (3) this may influence which crops are prioritized for the core ML demonstration in this project.

```python
import pandas as pd
import numpy as np

clean_df = pd.read_csv("../data/processed/clean_mandi_prices.csv")
clean_df['Price Date'] = pd.to_datetime(clean_df['Price Date'], format='%Y-%m-%d')
clean_df['Year'] = clean_df['Price Date'].dt.year
clean_df['Month'] = clean_df['Price Date'].dt.month

print(clean_df.shape)
print(clean_df.dtypes['Price Date'])
```

```python
volatility = clean_df.groupby('Commodity')['Modal_Price'].agg(['mean', 'std'])
volatility['CV'] = (volatility['std'] / volatility['mean']) * 100
volatility = volatility.sort_values('CV', ascending=False)
volatility.round(2)
```

```python
monthly_crop_prices = clean_df.groupby(
    ['Commodity', 'Year', 'Month']
)['Modal_Price'].mean().reset_index()

monthly_crop_prices.columns = ['Commodity', 'Year', 'Month', 'Avg_Modal_Price']
monthly_crop_prices = monthly_crop_prices.sort_values(
    ['Commodity', 'Year', 'Month']
).reset_index(drop=True)

print(monthly_crop_prices.shape)
monthly_crop_prices.head(10)
```

```python
import os
os.makedirs("../data/processed", exist_ok=True)

# File 1 — already built above
monthly_crop_prices.to_csv(
    "../data/processed/monthly_crop_prices.csv", index=False
)

# File 2 — crop summary
crop_summary = clean_df.groupby('Commodity')['Modal_Price'].agg(
    ['mean', 'std', 'min', 'max', 'count']
).round(2).reset_index()
crop_summary['CV'] = (crop_summary['std'] / crop_summary['mean'] * 100).round(2)
crop_summary.columns = [
    'Commodity', 'Mean_Price', 'Std_Price',
    'Min_Price', 'Max_Price', 'Record_Count', 'CV_Percent'
]
crop_summary.to_csv(
    "../data/processed/crop_summary.csv", index=False
)

# File 3 — district price summary
district_summary = clean_df.groupby(
    ['District Name', 'Commodity']
)['Modal_Price'].agg(['mean', 'std', 'count']).round(2).reset_index()
district_summary.columns = [
    'District', 'Commodity', 'Mean_Price', 'Std_Price', 'Record_Count'
]
district_summary = district_summary[
    district_summary['Record_Count'] >= 30
].reset_index(drop=True)
district_summary.to_csv(
    "../data/processed/district_price_summary.csv", index=False
)

print("Files saved successfully")
print("monthly_crop_prices:", monthly_crop_prices.shape)
print("crop_summary:", crop_summary.shape)
print("district_summary:", district_summary.shape)
```

## Day 4 Summary — Aggregation & Time-Series Preparation

### Setup
Loaded clean_mandi_prices.csv (188,080 rows × 10 cols), re-applied datetime 
conversion using format='%Y-%m-%d' (ISO format — pandas writes dates in this 
format when saving to CSV, resolving the M/D/Y ambiguity from Day 3). 
Extracted Year and Month as separate integer columns using `.dt.year` and 
`.dt.month`.

### Part 1 — Datetime understanding
- Confirmed Price Date dtype is datetime64[us] after conversion
- Extracted Year (2023/2024/2025) and Month (1-12) as features
- Understood why feature extraction matters: ML models need numeric columns, 
  not raw datetime objects — Month captures seasonality, Year captures 
  long-term trend

### Part 2 — GroupBy fundamentals
- Learned groupby analogy: sorting receipts into envelopes, then computing 
  per-envelope
- Average Modal_Price by crop: Tomato highest (₹3724), Potato lowest (₹1357)
- Discovered market-level groupby is misleading when mixing crops — 
  Shamli market showed ₹4201 "average" but had only 7 rows, all Onion, 
  making it statistically unreliable
- Correct approach: always groupby ['Market', 'Commodity'] together, 
  never just Market alone for price comparisons
- Many (Market, Commodity) combinations have only 1 row — market-level 
  estimates unreliable for sparse combinations; dashboard should use 
  district-level or show a minimum-data warning

### Part 3 — Monthly aggregation
- Computed monthly average Modal_Price per crop (combining years) and 
  per crop per year separately
- Discovered critical data limitation through investigation:
  - Rice: all data from 2025 only (Apr-Jun) — 3 months, single year
  - Tomato: all data from 2023 only (Jun-Nov) — 6 months, single year
  - Onion/Potato: full coverage across all 3 years (boundary-effect gaps 
    only — e.g., no Jan 2023 because dataset starts Jun 2023)
  - Wheat: 9 months across years
- Conclusion: Rice and Tomato coverage gaps are dataset construction 
  artifacts, NOT real seasonality. Seasonality analysis reliable only 
  for Onion, Potato, Wheat.

### Part 4 — Pivot tables
- Built mean pivot (Commodity × Month → avg price) and count pivot 
  (Commodity × Month → record count)
- Count pivot is a "confidence map" — cells with few records 
  (e.g., Tomato November: 143 records) are less reliable than 
  cells with thousands
- Key seasonal patterns visible:
  - Onion: low in June (₹1192), peaks November (₹3696) — 3x swing
  - Tomato: dramatic July spike (₹6542 vs June's ₹1659) — monsoon effect
  - Wheat: nearly flat (₹2251-₹2509) — MSP price support confirmed

### Part 5 — Volatility analysis
Computed CV (Coefficient of Variation = std/mean × 100) per crop:

| Crop    | Mean (₹) | Std (₹) | CV (%)  |
|---------|----------|---------|---------|
| Tomato  | 3724     | 2730    | 73.30   |
| Onion   | 2215     | 945     | 42.66   |
| Potato  | 1357     | 520     | 38.29   |
| Rice    | 3229     | 394     | 12.19   |
| Wheat   | 2353     | 131     | 5.55    |

- CV preferred over raw std for cross-crop comparison (adjusts for 
  price scale differences)
- Wheat's low CV confirms government MSP intervention
- Tomato's CV based on single year only — treat as indicative, not definitive
- Implications for Day 16 ranking engine: CV will feed directly into 
  "risk/stability" scoring component
